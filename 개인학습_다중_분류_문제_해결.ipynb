```python
import pandas as pd
df = pd.read_csv('/content/iris.csv', names=["sepal_length", 'sepal_width', "petal_length", "petal_width", "species"])
print(df.head())
```

```python
# 그래프 출력
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue="species")
plt.show()
```

- 품종별로 크기와 너비 차이가 있음을 알 수 있다. 

### 원핫인코딩

```python
# 원핫인코딩
dataset = df.values
dataset
```

```python
# 전체 행을 가져오고 열 인덱스는 0-3까지만 출력시킴
X = dataset[:,0:4].astype('float')
X
```

```python
y = dataset[:,4] # 전체 행과 3번째 인덱스를 가지고 오라는 의미 
y
```

2) LabelEncoder : y값은 문자열 -> 숫자로 변경

```python
# y값은 문자열 -> 숫자 형태로 클래스를 변경해줘야 함
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
e.fit(y)
```

```python
e
```

```python
# 변환 Transform (문자->숫자형)
Y = e.transform(y)
Y
```

- iris-setosa = 0
- iris-versicolor = 1
- iris-virginica = 2

3) 활성화 함수를 적용시키려면 y값이 0과 1로 이루어져야함
-> tf.keras.utils.categorical()함수 사용

```python
from keras.utils import np_utils

Y_encoded = np_utils.to_categorical(Y)
```

```python
Y_encoded # 원-핫 인코딩화 시킴
```

### 소프트맥스
- 총합이 1인 형태로 바꿔서 계산해주는 함수
- 큰 값이 두드러지고 작은 값은 더 작아짐
- input_dim = 몇개의 값을 가져올 것인가


```python
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
model = Sequential()
# 16개의 노드에 4개의 값을 전달해줌
model.add(Dense(16, input_dim=4, activation='relu'))
# 최종 출력 값은 3개 중 하나여야 하므로 노드 수를 3으로 설정
model.add(Dense(3, activation='softmax'))
```

### 모델 compile(실행)
- 다중분류는 categorical_crossentropy (오차함수) 사용
- 최적화 함수로 adam 사용
- 전체 샘플이 50회 반복될 때까지 한 번에 입력되는 값은 1개

```python
# model compile
model.compile(loss='categorical_crossentropy',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```python
# 모델 실행 fit
model.fit(X, Y_encoded, epochs=50, batch_size=1)
```

- 정확도 : 97%
